# creating new features from existing ones

```python
data["fours"] = (data["total_runs"]==4).astype(int)
data["sixes"] = (data["total_runs"]==6).astype(int)
```

```python
data['total_runs'] = data.groupby("id")['total_runs'].cumsum()
data['extra_runs'] = data.groupby("id")['extra_runs'].cumsum()
data['is_wicket'] = data.groupby("id")['is_wicket'].cumsum()
data['fours'] = data.groupby("id")['fours'].cumsum()
data['sixes'] = data.groupby("id")['sixes'].cumsum()
```

```python
# data = data.merge(group[["id","over","run_rate"]],on=["id","over"],how='left')
data['run_rate']= (data['total_runs']/(data['ball']+data['over']*6))*6
```

```python
data.reset_index(drop=True)
```

# creating a balls_left column

```python
data["extras_type"] = (data["extras_type"].isin(['wides','noballs'])==True).astype(int)
indx = data[data.extras_type == 1].index
for i in indx:
    data['ball'][i]=0
data["ball"] = (data["ball"]!=0).astype(int)
data["ball"] = data.groupby(["id","over"])["ball"].cumsum()
for i in indx:
    data['ball'][i]=0
for i in data.index:
    if data.ball[i]==0:
        data.ball[i]=data.ball[i-1]
data["balls_left"] = 120 - (data["over"]*6 + data['ball'])
```
