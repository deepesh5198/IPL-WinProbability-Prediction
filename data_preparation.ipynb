```python
import pandas as pd
import numpy as np
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
df = pd.read_csv("IPL Ball-by-Ball 2008-2020.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
new_df = df.groupby(["id","inning"]).agg({"total_runs": sum,
                                          "over": lambda x: max(x)+1,
                                          "is_wicket":sum,
                                          "batting_team": lambda x: x.iloc[0],
                                         "bowling_team": lambda x: x.iloc[0]}).reset_index()
```

```python
new_df
```

```python
part1 = new_df[new_df["inning"] == 1]
part2 = new_df[new_df["inning"] == 2]
```

```python
display(part1, part2)
```

```python
set(part1['id']) - set(part2['id'])
```

```python
type(part1)
```

```python
part1.drop(part1[part1.id.isin([501265,829763])].index, inplace=True)
```

```python
dataset = part1.drop("inning", axis=1)
dataset.head(1)
```

```python
dataset = dataset.rename(columns = {"total_runs": "target",
                         "over": "overs_played",
                         "is_wicket": "total_wickets",
                         "batting_team": "team1",
                         "bowling_team": "team2"})
```

```python
dataset.head(1)
```

```python
dataset["run_rate"] = dataset["target"]/dataset["overs_played"]
```

```python
dataset.head(1)
```

```python
df2 = pd.read_csv("IPL Matches 2008-2020.csv")
```

```python
df2.head(5)
```

```python
dataset = dataset.merge(df2[["id","venue","toss_decision","winner"]], on="id", how="left")
```

```python
dataset
```

```python
dataset.isna().sum()
```

```python
dataset["winner"] = (dataset['winner'] == dataset['team2']).astype(int)
```

```python
dataset["team2_won_toss?"] = (dataset['toss_decision'] == 'field').astype(int)
```

```python
dataset["team1_all_out?"] = (dataset['total_wickets'] == 10).astype(int)
```

```python
dataset = dataset.drop("id", axis = 1)
```
